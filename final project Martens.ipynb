```python
# ANDREW MARTENS
# COSC 210
# FINAL PROJECT
# CONNECT 4
# 1/26/2023


def spot_check(game_board, y):
    """Checks if column is full. If not full identifies first open spot in array"""
    ROW = 6
    global x
    # If the column has an open space
    if game_board[ROW - 1][y] == 0:
        
        # For each spot in the row it detects the first open spot
        for r in range(ROW):
            if game_board[r][y] == 0:
                x = r
                return True
    else:
        return False
    
# Method to place a piece on the game board
def place_piece(game_board, x, y, piece):
    """Places 1 or 2 in the game board array"""
    game_board[y][x] = piece # piece will equal a 1 or 2
    
# Method to check if someone won the game
def win(game_board, piece):
    """Checks each possible combination of connect 4 and returns true if connect 4 is detected"""
    # Checks for horizontal win from left to right
    for c in range(COLUMN - 3):
        for r in range(ROW):
            if game_board[r][c] == piece and game_board[r][c+1] == piece and game_board[r][c+2] == piece and game_board[r][c+3] == piece:
                return True
    # Checks for vertical win from bottom to top
    for c in range(COLUMN):
        for r in range(ROW - 3):
            if game_board[r][c] == piece and game_board[r+1][c] == piece and game_board[r+2][c] == piece and game_board[r+3][c] == piece:
                return True
    # Checks for upwards diagnal win
    for c in range(COLUMN - 3):
        for r in range(ROW - 3):
            if game_board[r][c] == piece and game_board[r+1][c+1] == piece and game_board[r+2][c+2] == piece and game_board[r+3][c+3] == piece:
                return True
    # Checks for a downwards diagnal win 
    for c in range(COLUMN - 3):
        for r in range(3, ROW):
            if game_board[r][c] == piece and game_board[r-1][c+1] == piece and game_board[r-2][c+2] == piece and game_board[r-3][c+3] == piece:
                return True    
    

def display(x_pos, turn, game_board, screen):
    """Displays the images for the connect 4 game"""
    # Makes a blank screen
    screen.fill(WHITE)
    
    # Changes floating chip to either red or yellow based on the turn
    if turn == 0:
        red_chip = pygame.draw.circle(screen, RED, (x_pos, y_pos), 50)
    else:
        yellow_chip = pygame.draw.circle(screen, YELLOW, (x_pos, y_pos), 50)
    
    # Displays the placed chips
    for r in range(ROW):
        for c in range(COLUMN):
            if game_board[r][c] == 1:
                pygame.draw.circle(screen, RED, ((50+c*100), (650-r*100)), 50)
            elif game_board[r][c] == 2:
                pygame.draw.circle(screen, YELLOW, ((50+c*100), (650-r*100)), 50)
                
    # Displays the connect 4 board
    screen.blit(board_display, (0, 100))
    
    # Refreshes the image
    pygame.display.flip()

# method to check for a draw game state
def draw(game_board):
    # Looks for any open spot and returns false if an empty spot is found
    for row in game_board:
        for spot in row:
            if spot == 0:
                return False
    return True
    
    
    

    
def end_of_game(game_board, win_text, play_again, red_win, yellow_win, draw_text, screen):
    """Displays the board and text for when the game ends"""
    # Blank screen
    screen.fill(WHITE)
    
    # Displays chips and game board
    for r in range(ROW):
        for c in range(COLUMN):
            if game_board[r][c] == 1:
                pygame.draw.circle(screen, RED, ((50+c*100), (650-r*100)), 50)
            elif game_board[r][c] == 2:
                pygame.draw.circle(screen, YELLOW, ((50+c*100), (650-r*100)), 50)
    screen.blit(board_display, (0, 100))
    
    # If red wins
    if win_text == 1:
        
        # Displays RED WINS!!!
        screen.blit(red_win, (250, 0))
        # Displays 'press space to play again'
        screen.blit(play_again, (125, 35))
        
    elif win_text == 2:
        
        # Displays YELLOW WINS!!!
        screen.blit(yellow_win, (250, 0))
        # Displays 'press space to play again'
        screen.blit(play_again, (125, 35))

    elif win_text == 0:
        # displays "DRAW!!"
        screen.blit(draw_text, (250, 0))
        screen.blit(play_again, (125,35))
        
    # Refreshes image
    pygame.display.flip()
    
    
def main():
    
    pygame.init()
    

    game_board = np.zeros((ROW,COLUMN))
    
    # Sets values for red and yellow pieces
    red_piece = 1
    yellow_piece = 2
    
    # Variables run and game keep the game running
    run = True
    game = True
    turn = 0
    
    # Variables to position the floating chip
    x_pos = 50
    y_pos = 50
    chip_radius = 50

    # Sets variable for the screen
    screen = pygame.display.set_mode((WIDTH, HEIGHT))
    # Writes the caption name
    pygame.display.set_caption('CONNECT 4')
    
    
    # Importing font 'monospace' with size of 30
    textfont = pygame.font.SysFont('monospace', 30)
    
    # Assigns text for when red or yellow wins or if there is a draw
    red_win = textfont.render("RED WINS!!", 1, BLACK)
    yellow_win = textfont.render("YELLOW WINS!!", 1, BLACK)
    draw_text = textfont.render("TIE GAME!!", 1, BLACK) 
    
    # Tells user what to press to play again
    play_again = textfont.render("PRESS SPACE TO PLAY AGAIN", 1, BLACK)
    
    
    
    while run:
        
        while game:
            
            # Player 1's turn
            if turn == 0:
                
                # For input, pygame gets it runs this if statement
                for event in pygame.event.get():
                    
                    # If user exits out of the screen it ends the game
                    if event.type == pygame.QUIT:
                        run = False
                        game = False
                        win_text = 3
                    
                    if event.type == pygame.KEYDOWN:
                        
                        # If user presses the left arrow it moves the floating chip left one column
                        if event.key == pygame.K_LEFT:
                            x_pos -= 100
                            if x_pos < 0:
                                x_pos = 0 + chip_radius  # Acts as boarder
                                
                        # If user presses the right arrow it moves the floating chip right one column
                        if event.key == pygame.K_RIGHT:
                            x_pos += 100
                            if x_pos > WIDTH:
                                x_pos = WIDTH - chip_radius  # Acts as boarder
                        
                        # If user presses the down arrow this code detects the location of the chip
                        # and assigns y as the column
                        if event.key == pygame.K_DOWN:
                            if x_pos == 50:
                                y = 0
                            elif x_pos == 150:
                                y = 1
                            elif x_pos == 250:
                                y = 2
                            elif x_pos == 350:
                                y = 3
                            elif x_pos == 450:
                                y = 4
                            elif x_pos == 550:
                                y = 5
                            elif x_pos == 650:
                                y = 6
                            
                            if not spot_check(game_board, y):  # See above for details
                                continue
                            place_piece(game_board, y, x, red_piece)  # See above for details
                            winner = win(game_board, red_piece)  # See above for details
                            tie = draw(game_board)
                            
                            # If winner detected ends the game and displays the end game text
                            if winner:
                                win_text = 1
                                game = False
                            elif tie:
                                win_text = 0
                                game = False
                            

                            
                            # Changes turn to skip over player 2's code
                            turn = 3
                            
            # Player 2's turn
            elif turn == 1:
            
                for event in pygame.event.get():
                    
                    # If user exits out of the screen it ends the game
                    if event.type == pygame.QUIT:
                        run = False
                        game = False
                        win_text = 3
                        
                    # If user presses the left arrow it moves the floating chip left one column
                    if event.type == pygame.KEYDOWN:
                        if event.key == pygame.K_LEFT:
                            x_pos -= 100
                            if x_pos < 0:
                                x_pos = 0 + chip_radius  # Acts as boarder
                        
                        # If user presses the right arrow it moves the floating chip right one column
                        if event.key == pygame.K_RIGHT:
                            x_pos += 100
                            if x_pos > WIDTH:
                                x_pos = WIDTH - chip_radius  # Acts as boarder
                        
                        # If user presses the down arrow this code detects the location of the chip
                        # and assigns y as the column
                        if event.key == pygame.K_DOWN:
                            if x_pos == 50:
                                y = 0
                            elif x_pos == 150:
                                y = 1
                            elif x_pos == 250:
                                y = 2
                            elif x_pos == 350:
                                y = 3
                            elif x_pos == 450:
                                y = 4
                            elif x_pos == 550:
                                y = 5
                            elif x_pos == 650:
                                y = 6
            
            
                            if not spot_check(game_board, y):  # See fun above for details
                                continue
                            place_piece(game_board, y, x, yellow_piece)  # See fun above for details
                            winner = win(game_board, yellow_piece)  # See fun above for details
                            tie = draw(game_board)
                        
                            # If winner detected ends the game and displays the end game text
                            if winner:
                                win_text = 2
                                game = False
                            elif tie:
                                win_text = 0
                                game = False
                            
                            # Changes turn back to player 1
                            turn = 0    
            
            # An odd turn = 1, even turn = 0
            turn = turn%2  
            
            # Runs the display funtion
            display(x_pos, turn, game_board, screen)
            
        # After someone wins:
        # Quits game if user exits
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                run = False
                
            # If user presses space resets board, turn, and sets game to True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE:
                    game_board = np.zeros((ROW,COLUMN))
                    turn = 0
                    game = True
                    
        # Displays the end of game text
        end_of_game(game_board, win_text, play_again, red_win, yellow_win, draw_text, screen)
        
    pygame.quit()

```
